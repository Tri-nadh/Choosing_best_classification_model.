**Data Preparation and Splitting**

```python
import pandas as pd
data = pd.read_csv('fertility_Diagnosis.txt',header=None)
X = data.iloc[:,:9]
Y = data.iloc[:,9]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=2,test_size=0.2)
```

**Classification Algorithms**

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
model_nb = GaussianNB()
model_nb.fit(X,Y)
pred_nb = model_nb.predict(x_test)
```

```python
#Decision_tree
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier()
model_tree.fit(X,Y)
pred_tree = model_tree.predict(x_test)
```

```python
#Naive Bayes
from sklearn.svm import SVC
model_svc = SVC()
model_svc.fit(X,Y)
pred_svc = model_svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
scores= [accuracy_score(y_test,pred_nb),accuracy_score(y_test,pred_tree),accuracy_score(y_test,pred_svc)]
print(scores)
```

**Comparing performance**

```python
from matplotlib import pyplot as plt
plt.plot(['GaussianNB','DecitionTree','SVC'],scores)
plt.xlabel('Clasifier')
plt.ylabel('Scores')
```

For this data DecitionTree Classifier is performing good
